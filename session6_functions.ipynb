### Let's create a function(with docstring)




```python
def is_even(num):
  """
  this is the docstring , this function returns if a given no , is odd or even
  input - any valid integer
  output - odd or  even
  created on - 28 july,25
  """

  if type(num) == int:
    if num % 2 == 0:
      return 'even'
    else:
      return 'odd'
  else:
    return 'please enter an integer'

```

```python
is_even(7)  #function calling
```

```python
for i in range(1,11):
  print(i,is_even(i))
```

```python
# Accessing The Documentation of a function

print(is_even.__doc__)

print(print.__doc__)

print(type.__doc__)
```

### Types of Arguments

- Default Argument
- Positional Argument
- Keyword Argument

```python
#default arguments =>  if a ko value na mile then a = 1, and b ko value na mile then b = 1 , default setting

def power(a=1,b=1):
  return a**b
```

```python
power(2,3)
```

```python
power(2) #here a = 1,b = 1
```

```python
power()
```

```python
#positional arguments

power(2,3)  #1st arg goes to 1st parameter, 2nd -2nd ......... very default setting
```

```python
#keyward argument

power(b=3,a=2)  #here we can pass the argumnets by the keywords
```

### `*args and **kwargs`
`*args and **kwargs` are special Python keywords that are used to pass the variable length of arguments to a function

```python
# *args
# allows us to pass a variable number of non-keyword arguments to a function.
```

```python
def multiply(a,b,c):
  return a*b*c
```

```python
multiply(2,3,4)
```

```python
# when the arguments or the inputs given by the customer is variable we dont not know how many will be given then we use the *args
```

```python
def multiply(*args):       #* lagiye deya manei python bujhe nicche je variable arguments/inputs
  product = 1
  for i in args:
    product = product*i
  print(args)
  return product
```

```python
multiply(1,2,4,4,56,6,1,1)
```
