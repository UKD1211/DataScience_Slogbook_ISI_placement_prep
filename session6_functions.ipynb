### `*args and **kwargs`
`*args and **kwargs` are special Python keywords that are used to pass the variable length of arguments to a function

```python
# *args
# allows us to pass a variable number of non-keyword arguments to a function.
```

```python
def multiply(a,b,c):
  return a*b*c
```

```python
multiply(2,3,4)
```

```python
# when the arguments or the inputs given by the customer is variable we dont not know how many will be given then we use the *args
```

```python
def multiply(*args):       #* lagiye deya manei python bujhe nicche je variable arguments/inputs
  product = 1
  for i in args:
    product = product*i
  print(args)
  return product
```

```python
multiply(1,2,4,4,56,6,1,1)
```

```python
# **kwargs
# **kwargs allows us to pass any number of keyword arguments.
# Keyword arguments mean that they contain a key-value pair, like a Python dictionary.
# if we want to send key-value pairs as functions then we can send **kwargs
```

```python
def display(**kwargs):
  for (key,value) in kwargs.items():
    print(key,'->',value)
```

```python
display(india = 'delhi',nepal = 'kathmandu',pakistan = 'islamabad')
```

##### Points to remember while using `*args and **kwargs`

- order of the arguments matter(normal -> `*args` -> `**kwargs`)
- The words args and kwargs are only a convention, you can use any name of your choice

### Lambda Function

A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.

image.png

```python
# lambda function has no name
```

```python
#x -> x^2
a = lambda x:x**2
a(2)
```

```python
# x,y -> x+y
b = lambda x,y: x+y
b(2,3)
```

#### Diff between lambda vs Normal Function

- No name
- lambda has no return value(infact,returns a function)
- lambda is written in 1 line
- not reusable

Then why use lambda functions?<br>
**They are used with HOF(Higher Oder Functions)**


```python
#check if a string has a

# normal function
def check(x):
  b = str(x)
  for i in b:
    if i == 'a'or i=='A':
      return "here is 'a'"
  return "here is no a"
```

```python
# check('lola')
# check(11)
m = check
m('soumya')
```

```python
def square(x):
  return x**2
                                 #this a normal function here
def list_new(L):
  L_new = L.split()
  result = []
  for i in L_new:
    result.append(square(int(i))) # Convert string to integer here
  return result
```

```python
print(list_new("1 2 3 4"))
```

```python
def square(x):
  return x**2

def cube(x):
    return x**3

# this tranform() is the HOF
def transform(f,L):
    output = []
    for i in L:
        output.append(f(i))   #here its a HOF , as we are giving the an function here as input
    print(output)
```
