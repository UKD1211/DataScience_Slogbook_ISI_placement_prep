```python
def sum_iterative(l):
  sum = 0
  for i in l:
    sum += i
  return sum
```

```python
def sum_recarsive(l):
  if len(l) == 0:
    return 0
  else:
    return (l[0] + sum_recarsive(l[1:]))
```

```python
# Factorial
def fact(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n*fact(n-1)
fact(5)
```

```python
MOD = 10**9 + 7

def count_good_string(n):
  even_count = (n+1)//2
  odd_count = n//2

  return (pow(5,even_count, MOD)*pow(4,odd_count,MOD)) % MOD
count_good_string(5)
```

```python
mod =  10**9 + 7

def powe(x,n,mod):
  if n == 0:
    return 1
  half = power(x,n//2,mod)
  result = (half*half) % mod
  if n%2 == 1:
    result = (result*x) % mod
  return result
```

```python
from typing import List

class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        result = []

        def doit(current: str, open_left:int, close_left:int):
          if open_left == 0 and close_left == 0:
            result.append(current)
            return
          if open_left>0:
            doit(current + '(', open_left-1,close_left)
          if close_left > open_left:
            doit(current + ')', open_left,close_left-1)
        doit('',n,n)
        return result
```

```python
d = {
    2: ["a", "b", "c"],
    3: ["d", "e", "f"],
    4: ["g", "h", "i"],
    5: ["j", "k", "l"],
    6: ["m", "n", "o"],
    7: ["p", "q", "r", "s"],
    8: ["t", "u", "v"],                      #This is only for upto 2 digits
    9: ["w", "x", "y", "z"]
}


def keypad(n):
  if len(list(str(n))) == 1:
    return d[int(n)]
  else:
    new_L =[]
    for i in d[int(list(str(n))[0])]:
      for j in d[int(list(str(n))[1])]:
        new_L.append(i + j)

    return new_L
```

```python
from typing import List

class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        phone_map = {
            '2': 'abc', '3': 'def', '4': 'ghi', '5': 'jkl',
            '6': 'mno', '7': 'pqrs', '8': 'tuv', '9': 'wxyz'
        }

        result = []

        def backtrack(index: int, current: str):
            # base case
            if index == len(digits):
                result.append(current)
                return

            for char in phone_map[digits[index]]:
                backtrack(index + 1, current + char)

        if digits:  # to avoid running when digits is empty
            backtrack(0, "")

        return result
```
