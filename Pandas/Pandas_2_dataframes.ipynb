```python
#write a function that can return the track record of 2 teams against each other
```

```python
ipl.head()
```

```python
x = ipl[(ipl['Team1'] == 'Royal Challengers Bangalore') & (ipl['Team2'] == 'Kolkata Knight Riders')][['Season',"Team1","Team2",'WinningTeam']]
```

```python
y = ipl[(ipl['Team2'] == 'Royal Challengers Bangalore') & (ipl['Team1'] == 'Kolkata Knight Riders')][['Season',"Team1","Team2",'WinningTeam']]
```

```python
pd.concat([x,y])
```

```python
def track(a:str,b:str):
  x = ipl[(ipl['Team1'] == a) & (ipl['Team2'] == b)][['Season',"Team1","Team2",'WinningTeam']]
  y = ipl[(ipl['Team2'] == a) & (ipl['Team1'] == b)][['Season',"Team1","Team2",'WinningTeam']]
  return pd.concat([x,y])
track('Lucknow Super Giants','Royal Challengers Bangalore')
```
